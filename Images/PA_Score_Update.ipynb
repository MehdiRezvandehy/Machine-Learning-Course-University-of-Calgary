# PA_Score (Likelihood)

Calculate payment arrangement score (likelihood) as the probability (between 0 and 1) to answer this question "**what is the likelihood if a customer take payment arrangement (PA)?**"

We propose two techniques 1- Poisson Distribution, 2- Sigmoid function (Modified):

## Poisson Distribution

In probability theory and statistics, the Poisson distribution is a discrete probability distribution that expresses the probability of a given number of events occurring in a **fixed interval of time (or space)** if these events occur with a **known constant mean rate**:

$\LARGE P(n)=\frac{\mu ^{n}e^{-\mu}}{n!}$

where $P(n)$ is predicted probability for $n$ number of occurrence, $\mu$ is the mean for number occurrence for a fixed interval of time

### Small example

A small company, **on average**, is receiving *12* customers per day ($\mu=12$). What is the probability if the company will receive exactly *8* ($n=8$) customers in one day?

$\large P(n=8)=\frac{\mu ^{n}e^{-\mu}}{n!}=\frac{12 ^{8}e^{-12}}{8!}=0.0655$

In our work we need to calculate probability if a customer is take payment arrangement? The question can be like this:
Customer A has taken 4 payment arrangements on average per 60 days of delinquency episode; what is the probability if Customer A will take payment arrangements (more than 1) per 30 days ($P(n>0)$)?

$\large \mu_{60}=4 \,\, \rightarrow \,\,\mu_{30}=2$ 

$\large P(n=0)=\frac{\mu ^{n}e^{-\mu}}{n!}=\frac{2 ^{0}e^{-2}}{2!}=0.067$

$\large P(n>0)=1.0-P(n=0)=1.0-0.067=0.93$


```python
def poisson (m_n_pa,n=0):
    import math
    P_0=((m_n_pa**(n))*np.exp(-m_n_pa))/(math.factorial(n))
    P=1.0-P_0
    return P
```

##  Sigmoid Function (Modified) 

### Sigmoid Function

[Sigmoid Function](https://en.wikipedia.org/wiki/Sigmoid_function#:~:text=4%20Applications-,Definition,refer%20to%20the%20same%20object)  is a S-shaped function that has a lot of application in Machine Learning (Logistic Regression) and Artificial Intelligence (activation function). The output of *Sigmoid Function* is always between 0 and 1. The Equation below shows *Sigmoid Function*:

$\large S(x)=\frac{1}{1+e^{-x}}$


```python
fig = plt.subplots(figsize=(7,3.5), dpi= 100, facecolor='w')

x = np.linspace(-5, 5, 100)
S = 1 / (1 + np.exp(-x))

plt.plot(x, S, "r-", linewidth=3, label=r"$S(t) = \frac{1}{1 + e^{-x}}$")
plt.plot([-5, 5], [0, 0], "k-")
plt.plot([-5, 5], [0.5, 0.5], "k--")
plt.plot([-5, 5], [1, 1], "k--")
plt.plot([0, 0], [-1.1, 1.1], "k-")
plt.axis([-5, 5, 0, 1.1])
plt.xlabel("t", fontsize=14)
plt.ylabel("Probability "+ r'($\bar{p}$)', fontsize=14)
plt.title("Sigmoid Function", fontsize=16)
plt.legend(loc=4, fontsize=14)
plt.yticks([0,0.25,0.5,0.75,1.0])

plt.show()
```

In this work, sigmoid function is applied on calculated ratio. x variable in sigmoid function is substituted with the $r$:

$\large r=(\frac{\sum \,n\_pa}{\sum \, length})\times30$:

where $r$ is the ratio for number of n_pa per one month

```python
font = {'size'   :10}
plt.rc('font', **font)
fig,ax = plt.subplots(figsize=(7,3.5), dpi= 110, facecolor='w')
plt.rcParams['agg.path.chunksize'] = 100000

Ratio_sort=np.arange(0,10,0.001)

def function(i,K=1):
    y=(2/(1 + np.exp(-K*i)))-1
    return y

S = function(Ratio_sort)
PA_Score=S
plt.plot(Ratio_sort, np.array(S), "r-", linewidth=3, label=r"$S(r) = \frac{1-e^{-r}}{1 + e^{-r}}$")
plt.axis([0,8, 0, 1.01])
plt.xlabel(r"$r=(\frac{\sum \,n\_pa}{\sum \, length})\times30$", fontsize=14)
plt.ylabel("PA Score", fontsize=14)
plt.title("Apply Ratio in Sigmoid Function to Calculate PA Score", fontsize=14)
plt.legend(loc=4, fontsize=14)
plt.yticks([0,0.25,0.5,0.75,1.0])
plt.grid(color='k', linestyle='-', linewidth=0.1)

plt.show()
```

```python
def histplt (val,bins,title,xlabl,xlimt,axt=None):
    """ Function for histogram plotting"""
    
    from matplotlib.offsetbox import AnchoredText
    from matplotlib.ticker import PercentFormatter
    
    ax1 = axt or plt.axes()
    val=np.array(val)
    plt.hist(val, bins=bins, weights=np.ones(len(val)) / len(val),ec='black')      
    n=len(val)
    Mean=np.nanmean(val)
    SD=np.sqrt(np.nanvar(val))
    Max=np.nanmax(val)
    Min=np.nanmin(val)
    P10=np.quantile(val, 0.1, axis=0)
    P50=np.quantile(val, 0.5, axis=0)
    P90=np.quantile(val, 0.9, axis=0)
    
    txt='n=%.0f \n $\mu$=%.3f  \n P10=%.2f  \n P50=%.2f   \n P90=%.2f \n Max=%.2f \n Min=%.2f'
    anchored_text = AnchoredText(txt %(n,Mean,P10,P50,P90,Max,Min), loc=1)
    ax1.add_artist(anchored_text)
    plt.title(title,fontsize=font['size']*1.35)   
    plt.xlabel(xlabl,fontsize=font['size'])            
    #plt.ylabel('Frequency',fontsize=font['size'])
    plt.xlim(xlimt)
    plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
    ax1.grid(linewidth='0.35')
    
    
def CrossPlot (x,y,xlabl,ylabl,axt=None):
    '''Cross plot between two variables'''
    from matplotlib.offsetbox import AnchoredText
    ax1 = axt or plt.axes()
    x=np.array(x)
    y=np.array(y)    
    n_x=len(x)
    Mean_x=np.mean(x)
    SD_x=np.sqrt(np.var(x)) 
    n_y=len(y)
    Mean_y=np.mean(y)
    SD_y=np.sqrt(np.var(y)) 
    corr=np.corrcoef(x,y)
    plt.plot(x,y,'ro',markersize=2,alpha=0.08)
    txt='$\\rho_{XY}$=%.3f \n $\mu_{X}$=%.2f \n '
    txt+=' $\mu_{Y}$=%.2f'
    anchored_text = AnchoredText(txt %(corr[1,0],Mean_x,Mean_y), loc=2)
    ax1.add_artist(anchored_text)
    plt.xlabel(xlabl,fontsize=font['size'])            
    plt.ylabel(ylabl,fontsize=font['size']) 
#    plt.plot([0,1],[0,1],'--') 
    plt.xlim(0,1)
    plt.ylim(0,1)  
    ax1.grid(linewidth='0.35')
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
def Shallow_ANN (input_dim,neurons=50,loss="sparse_categorical_crossentropy",activation="relu",Nout=4,
             metrics=['accuracy'],activation_out='softmax',init_mode=None,BatchOpt=False,dropout_rate=False):
    """ Function to run Neural Network for different hyperparameters"""
    np.random.seed(42)
    tf.random.set_seed(42)
    
    if(activation=='Leaky_ReLU'): activation = keras.layers.LeakyReLU(alpha=0.2)
        
    # create model
    model = keras.models.Sequential()
    
    # Input & Hidden Layer 1
    model.add(keras.layers.Dense(neurons,input_dim=input_dim, activation=activation, kernel_initializer=init_mode))
        
    # Hidden Layer 2
    model.add(keras.layers.Dense(neurons,activation=activation)) 
    if(BatchOpt): model.add(keras.layers.BatchNormalization())
        
    # Hidden Layer 3    
    model.add(keras.layers.Dense(neurons,activation=activation)) 
    if(BatchOpt): model.add(keras.layers.BatchNormalization())
    if(dropout_rate):  model.add(keras.layers.Dropout(dropout_rate))
    
    # Output Layer 
    model.add(keras.layers.Dense(Nout,activation=activation_out)) 
        
    # Compile model
    model.compile(optimizer='adam',loss=loss,metrics=metrics)
    return model
```

```python
import matplotlib.pyplot as plt

def plot(history):

    font = {'size'   : 10}
    plt.rc('font', **font)
    fig, ax=plt.subplots(figsize=(12, 4), dpi= 110, facecolor='w', edgecolor='k')
    
    ax1 = plt.subplot(1,2,1)
    history_dict = history.history
    loss_values = history_dict['loss']
    val_loss_values = history_dict['val_loss']
    
    epochs = range(1, len(loss_values) + 1)
    plt.plot(epochs, loss_values, 'bo', label='Training loss')          
    plt.plot(epochs, val_loss_values, 'r', label='Validation loss',linewidth=2)    
    plt.title('Training and Validation Loss',fontsize=14)
    plt.xlabel('Epochs (Early Stopping)',fontsize=12)
    plt.ylabel('Loss',fontsize=11)
    plt.legend(fontsize='12')
#    plt.ylim((0, 0.8))
    
    ax2 = plt.subplot(1,2,2)    
    history_dict = history.history
    loss_values = history_dict['accuracy']
    val_loss_values = history_dict['val_accuracy']
    epochs = range(1, len(loss_values) + 1)
    ax2.plot(epochs, loss_values, 'ro', label='Training accuracy')          
    ax2.plot(epochs, val_loss_values, 'b', label='Validation accuracy',linewidth=2)    
    plt.title('Training and Validation Accuracy',fontsize=14)
    plt.xlabel('Epochs (Early Stopping)',fontsize=12)
    plt.ylabel('Accuracy',fontsize=12)
    plt.legend(fontsize='12')
#    plt.ylim((0.8, 0.99))
    plt.show()
```

```python
def BS_ANN (x_train,y_train,x_Validation,y_Validation,neurons=50,loss="sparse_categorical_crossentropy",activation="relu"
            ,Nout=4,metrics=['accuracy'],activation_out='softmax',init_mode=None,BatchOpt=False,dropout_rate=False):
    """ Function to run Neural Network for different hyperparameters for Bootstraping"""
    np.random.seed(42)
    tf.random.set_seed(42)
    
    if(activation=='Leaky_ReLU'): activation = keras.layers.LeakyReLU(alpha=0.2)
        
    # create model
    model = keras.models.Sequential()
    
    # Input & Hidden Layer 1
    model.add(keras.layers.Dense(neurons,input_dim=np.array(x_train).shape[1], activation=activation, kernel_initializer=init_mode))
        
    # Hidden Layer 2
    model.add(keras.layers.Dense(neurons,activation=activation)) 
    if(BatchOpt): model.add(keras.layers.BatchNormalization())
        
    # Hidden Layer 3    
    model.add(keras.layers.Dense(neurons,activation=activation)) 
    if(BatchOpt): model.add(keras.layers.BatchNormalization())
    if(dropout_rate):  model.add(keras.layers.Dropout(dropout_rate))
    
    # Output Layer 
    model.add(keras.layers.Dense(Nout,activation=activation_out)) 
        
    # Compile model
    model.compile(optimizer='adam',loss=loss,metrics=metrics)
    
    # Early stopping to avoid overfitting
    monitor= tf.keras.callbacks.EarlyStopping(monitor='val_loss', min_delta=1e-3,patience=3, mode='auto')
    history=model.fit(x_train,y_train,batch_size=32,validation_data=
              (x_Validation,y_Validation),callbacks=[monitor],verbose=0,epochs=1000)
    history_dict = history.history
    loss_values = history_dict['loss']
    epochs = len(loss_values)-1    
    
    return model, epochs
```
